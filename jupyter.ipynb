```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
with open("scrims.json", "r", encoding = "utf-8") as f:
    data = json.load(f)

    print(type(data))
    print(data)

    df = pd.DataFrame(data)

    vitoria = 0 
    derrota = 0

    for vitorias in df["resultado"]:
        if vitorias == "vitória":
            vitoria =+ 1
        else:
            derrota =+ 1

            plt.pie([vitoria, derrota], labels = ["Vitórias", "Derrotas"], autopct = "%1.1f%%", colors = ["#66bb6a", "#ef5350"])

plt.title("Taxa de Vitória nas Scrims")
plt.show()
```
